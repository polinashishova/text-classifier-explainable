```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
prep_dir = os.path.join(os.path.dirname('__file__'), 'data', 'processed')

def load_data(path, file_name):
    data_list = []
    with open(os.path.join(path, file_name), 'r', encoding='utf-8') as data_file:
        for text in data_file.readlines():
            data_list.append(text.strip())
    return data_list

X_train_cleaned = load_data(prep_dir, 'cleaned-train.txt')
X_test_cleaned = load_data(prep_dir, 'cleaned-test.txt')

y_train = [int(label) for label in load_data(prep_dir, 'train-labels.txt')]
y_test = [int(label) for label in load_data(prep_dir, 'test-labels.txt')]
```
